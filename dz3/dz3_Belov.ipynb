```python
import numpy as np
import matplotlib.pyplot as plt
```

Взяли сигнал, как сумму синусов, отобразили его

```python
t = np.arange(0,10,0.001)
s = np.sin(2*np.pi*t)+np.sin(2*np.pi*t/5-np.pi/8)
plt.plot(t,s)
```

Добавили шум:

```python
n = np.random.uniform(np.min(s)/3, np.max(s)/3, len(t))
sn = s+n
plt.plot(t,sn)
plt.plot(t,s)
```

```python
print(np.min(s)/10, np.max(s)/10, len(t))
```

```python
fil = np.ones(100)

s_for_conv = np.ones(len(s)+2*len(fil)//2-1)
s_for_conv[len(fil)//2:len(fil)//2+len(s)] = sn
print(len(s_for_conv))
```

```python
s_matrix = []
for i in range(len(s)):
    s_matrix.append(s_for_conv[i:i+len(fil)])
print(len(s_matrix))
s_matrix = np.array(s_matrix)
print(s_matrix.shape)
```

```python
c = np.linalg.lstsq(s_matrix, s)
print(c[0])
```

```python
s_filt = s_matrix @ c[0]#[::-1]
plt.plot(t, sn, label = 'sn')
plt.plot(t, s_filt, label = 's_filt')
plt.legend()
```

```python
def filtering(signal_mat, filter_len, coeff):
    filt = np.zeros(filter_len)
    filtered_signal = np.array([])
    for j in range(len(signal_mat)):
        e = signal_mat[j] - np.dot(filt, signal_mat[j])
        filtered_signal = np.append(filtered_signal, np.dot(filt, signal_mat[j]))
        A = np.tile(signal_mat[j], ((len(filt),1))).T
        temp = coeff*A
        filt = filt + np.dot(temp, e)
    return filtered_signal
```

```python
def plot(coeff):
    filt_signal = filtering(s_matrix, 100, coeff)
    plt.plot(sn, label = 'Исходный зашумлённый сигнал')
    plt.plot(filt_signal, label = 'Отфильтрованный сигнал')
    plt.legend()
```
